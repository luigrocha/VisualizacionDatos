```r
covidT %>% 
  mutate(Date_reported = ymd(Date_reported)) %>% # coerce column from character to date
  group_by(Country) %>% 
  summarize(max_date = max(Date_reported))
```

```r
covidT %>% group_by(Date_reported) %>% filter(Date_reported == '2022-04-04')
```

```r
df2 <- covidT
#Creo 3 filtros con la maxima fecha por año 2020
tmp1 <- df2 %>% group_by(Date_reported) %>%
filter(Date_reported == '2020-12-31')
#Creo 3 filtros con la maxima fecha por año 2021
tmp2 <- df2 %>% group_by(Date_reported) %>%
filter(Date_reported == '2021-12-31')
#Creo 3 filtros con la maxima fecha por año 2022
tmp3 <- df2 %>% group_by(Date_reported) %>%
filter(Date_reported == '2022-04-04')
#combino los 3 dataframes
TotalesbyYear = rbind(tmp1,tmp2)
TotalesbyYear = rbind(TotalesbyYear,tmp3)
#Dataframe por año 
TotalesbyYear
```

```r
TotalesbyYear$Year = substr(TotalesbyYear$Date_reported,1,4)
head(TotalesbyYear)
```

```r
#2.- De la información provista en el archivo desarrolle
#un diagrama barras apiladas de los casos por año por región usando la librería NVD3
#https://covid19.who.int/WHO-COVID-19-global-data.csv
#grafico dinamico 1 Casos por año y region
c1<- nPlot(Cumulative_cases ~ Year, group = 'WHO_region',
           data = subset(TotalesbyYear),
           type = 'multiBarChart')

c1
```

```r
c1$print('Covid-1-Contagios')
```

```r
#El mismo grafico con  barras horizontales
#grafico dinamico 2 Muertes por año y region
c2 <- nPlot(Cumulative_deaths ~ Year, group = "WHO_region",
            data = TotalesbyYear, type = "multiBarHorizontalChart")

c2
```

```r
c2$print('Covid-2-Muertes')
```
