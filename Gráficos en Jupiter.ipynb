```r
data(HairEyeColor)
hairEye <- as.data.frame(HairEyeColor)
data(iris)
names(iris)
n1<- nPlot(Freq ~ Hair, group = 'Eye',
           data = subset(hairEye, Sex == "Female"),
           type = 'multiBarChart')
n1
```

```r
#con print podemos generar/ver  el codigo  y copiarlo si se desea incrustar en una pagina 
n1$print("chart1") #chart1 es el nombre de la seccion del html

#El mismo grafico con  barras horizontales
n2 <- nPlot(Freq ~ Hair, group = "Eye", data = hairEye, type = "multiBarHorizontalChart")
n2$print("chart2")
#n2
```
